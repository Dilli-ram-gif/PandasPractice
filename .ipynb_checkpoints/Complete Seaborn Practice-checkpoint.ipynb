### Distribution Plots

#### Distributoin plot

```python
sns.displot(crash_df['not_distracted'], kde=True)
```

```python
for col in crash_df.columns[:len(crash_df.columns)-1]:
    sns.distplot(crash_df[col], kde=True)
```

```python
sns.displot(crash_df['not_distracted'], kde=True, bins=25)
```

#### Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

### Categorical Plots

#### Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data= tips_df)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
```

#### Count plot

```python
sns.countplot(x='sex', data=tips_df)
```

#### Box Plot

```python
sns.boxplot(x='day', y="total_bill", data=tips_df, hue= 'sex')
```

```python
sns.boxplot(x='day', y="total_bill", data=tips_df, hue= 'sex')
plt.legend(loc=0)
```

#### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

#### Strip Plot

```python
plt.figure(figsize=(8,6))
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

#### Swarm Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
```

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot')
plt.legend(loc=0)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot_table(index='month', columns='year', values='passengers')
```
