# Merge pandas dataframes- merge() function

```python
df1 = pd.DataFrame({
    'employee': ['Julia', 'Marie', 'Adam', 'Nicole'],
    'department': ['Data Science', 'Web Development', 'Data Science', 'Cyber Security'],
    'Year': ['2005', '2008', '2011', '2002']
})


df2 = pd.DataFrame({
    'employee': ['Nicole', 'Adam', 'Julia', 'Marie'],
    'country': ['Canada', 'England', 'USA', 'Germany'],
    'salary': ['22000', '16000', '20000', '17500']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='employee')
```

```python
pd.merge(df1,df2,on='employee', how='left')
```

```python
pd.merge(df1,df2,on='employee', how='right')
```

# Merge Pandas DataFrames-2

```python
df1 = pd.DataFrame({
    'employee': ['Julia', 'Marie', 'Adam', 'Nicole'],
    'department': ['Data Science', 'Web Development', 'Data Science', 'Cyber Security'],
    'Year': ['2005', '2008', '2011', '2002']
})


df2 = pd.DataFrame({
    'employee': ['Nicole', 'Adam', 'Julia', 'Marie'],
    'country': ['Canada', 'England', 'USA', 'Germany'],
    'salary': ['22000', '16000', '20000', '17500']
})
```

```python
df3 = pd.DataFrame({
    'manager': ['Abraham', 'Joseph', 'Kayra'],
    'department': ['Web Development', 'Cyber Security', 'Data Science']
    })
```

```python
df3
```

```python
df4 = pd.merge(df1, df2)
```

```python
df4
```

```python
df5 = pd.merge(df4, df3, on='department', how='right')
```
