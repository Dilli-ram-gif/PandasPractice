```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer

# as_frame = True, means we are loading the breast_dataset as dataframe
cancer = load_breast_cancer(as_frame=True)
df = cancer.frame
```

```python
# check the shape of the original data frame
print("The shape of the dataframe is ", df.shape)
```

```python
X = df[cancer['feature_names']]
print("The input dataframe shape is ", X.shape)
```

### Standardization

```python
# calculate the mean 
X_mean = X.mean()
X_std = X.std()
```

```python
# X_mean
# X_std

Z = (X - X_mean)/X_std
```

### Co-variance Matrix

```python
# Covariance
c = Z.cov()

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,12),)
sns.heatmap(c, annot=True, cmap='viridis')
plt.show()
```

### Eigen Values and Eigen Vectors

```python
# Calculate the eigen values and eigen vectors for our featrue space to identify principal components
```

```python
# tuple unpacking of eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(c)
print("The eigenvalues are: ", eigenvalues)
print("The eigenvectors are: ", eigenvectors)
```

```python
# shape of the eigenvalues and eigenvectos
print("The shape of the eigenvlaues is ", eigenvalues.shape)
print("The shape of the eigenvectors is ", eigenvectors.shape)
```

### Sort the eigenvalues in descending order

```python
# Index the eigenvalues in descending order
idx = eigenvalues.argsort()[::-1]

# Sort the eigenvalues in descending order
eigenvalues = eigenvalues[idx]

# Sort the corresponding eigenvectors in descending order
eigenvectors = eigenvectors[:, idx]
```

### Explained Variance
Explained variance is the term that gives us an idea of the amount of the total variance which has been retained by selecting the principal components instead of the original feature space.

```python
explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
explained_var
```

### Determine the number of Principal Components

```python
n_components = np.argmax(explained_var >= 0.5) + 1
n_components
```

### Project the Data onto the Selected Principal Components
- Find the projection matrix, It is a matrix of eigenvectors corresponding to the largest eigenvalues of the covariance matrix of the data. it projects the high-dimensional dataset onto a lower-dimensional subspace
- The eigenvectors of the covariance matrix of the data are referred to as the principal axes of the data, and the projection of the data instances onto these principal axes are called the principal components. 

```python
# PCA component or unit matrix
u = eigenvectors[:, :n_components-1]
pca_component =  pd.DataFrame(u, index=cancer['feature_names'], columns=['PC1, PC2'])
```

```python
# Plotting heatmap
plt.figure(figsize = (5,7))
sns.heatmap(pca_component, annot=True, cmap = 'viridis')
plt.title("PCA Component")
plt.show()
```

```python
# Matrix multiplication or dot Product
Z_pca = Z @ pca_component

Z_pca = pd.DataFrame(Z_pca.values,
					columns = ['PCA1','PCA2']
							)
Z_pca.head()
```

```python
# Importing PCA
from sklearn.decomposition import PCA

# Let's say, components = 2
pca = PCA(n_components=2)
pca.fit(Z)
x_pca = pca.transform(Z)

# Create the dataframe
df_pca1 = pd.DataFrame(x_pca,
					columns=['PC{}'.
					format(i+1)
						for i in range(n_components)])
df_pca1.head()
```
